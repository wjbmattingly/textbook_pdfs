```python
from PyPDF2 import PdfFileMerger
```

```python
import glob
```

```python
files = glob.glob("data/*.pdf")
```

```python
print (files)
```

```python
merger = PdfFileMerger()
```

```python
for file in files:
    merger.append(file)
merger.write("merged.pdf")
merger.close()
```
